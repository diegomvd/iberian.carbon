```python
import geopandas as gpd


spain_ccaa = gpd.read_file("SpainPolygon/gadm41_ESP_1.shp")
spain_ccaa = spain_ccaa.drop( [12,13] ).reset_index()
spain_ccaa
```

```python
spain_ccaa = spain_ccaa.dissolve()
spain_ccaa
```

```python
worldcover_grid = gpd.read_file("WorldCoverGrid/esa_worldcover_grid_composites.fgb")
```

```python
tiles_continental_spain = worldcover_grid[worldcover_grid.overlaps(spain_ccaa.geometry[0])]
```

```python
tiles_continental_spain
```

```python
import numpy as np
tiles_urls = np.array(tiles_continental_spain.drop(columns=["geometry","tile"]).values.tolist()).flatten().tolist()
tiles_urls
```

```python
import re
from pathlib import Path
from urllib.request import urlretrieve
import boto3
import botocore



s3 = boto3.resource('s3')

for url in tiles_urls:
    if 'esa-worldcover-s1' in url:
        bucket = re.findall('(s3://.*/)ESA',url)[0]
    if 'esa-worldcover-s2' in url:   
        bucket = re.findall('(s3://.*/)ESA',url)[0] 
    print(bucket)

    key = re.findall('{}(.*)'.format(bucket),url)[0]
    print(key)

    name = re.findall('(ESA_.*)',url)[0]
    if '2020' in name:
        path = "WorldCoverComposites/2020/{}".format(name)
    if '2021' in name:
        path = "WorldCoverComposites/2021/{}".format(name)    
    if not Path(path).exists():
        try:
            s3.Bucket(bucket).download_file(key, path)
        except botocore.exceptions.ClientError as e:
            if e.response['Error']['Code'] == "404":
                print("The object does not exist.")
            else:
                raise
```
