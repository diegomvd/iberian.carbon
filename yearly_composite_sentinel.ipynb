```python
import openeo

connection = openeo.connect("openeo.dataspace.copernicus.eu")

# List collections available on the openEO back-end
connection.list_collection_ids()
```

```python
# Get detailed metadata of a certain collection
connection.describe_collection("SENTINEL2_L2A")
```

```python
connection.describe_collection("SENTINEL1_GRD")
```

```python
connection.authenticate_oidc()
```

```python
#Iberian Peninsula
spatial_extent = {"west": -9.37, "south": 35.26, "east": 4.39, "north": 43.82}
#Quarter Iberian Peninsula
spatial_extent = {"west": 1.95, "south": 42, "east": 2.05, "north": 42.1}
spatial_extent
```

```python
datacube_sentinel2 = connection.load_collection(
    "SENTINEL2_L2A",
    spatial_extent=spatial_extent,
    temporal_extent = ["2019-01-01", "2022-01-01"],
    bands = ['B02', 'B03', 'B04'],
    # bands=['B01', 'B02', 'B03', 'B04', 'B05', 'B06', 'B07', 'B08', 'B8A', 'B09', 'B11', 'B12'],
    max_cloud_cover=85,
)
```

```python
composite_datacube = datacube_sentinel2.aggregate_temporal_period("year","mean")
```

```python
composite_datacube = composite_datacube.save_result(format="GTiff")
job = composite_datacube.create_job(title="Sentinel2-YearlyComposite-2017-2019")
```

```python
job_id = job.job_id
job
```

```python
job.start()
```

```python
job.status()
```
